```python
data1 = pd.read_csv("../../Documents/Data/Project/시도별_1인당_지역내총생산__지역총소득__개인소득_20210520104631.csv"
                   ,encoding="euc-kr", index_col=0)
data2 = pd.read_csv("../../Documents/Data/Project/유형별_매매가격지수_20210520105530.csv"
                   ,encoding="euc-kr", index_col=1)
data3 = pd.read_excel("../../Documents/Data/Project/행정구역_시군구_별__성별_인구수_20210520105140.xlsx"
                   )
con = lambda x : x.split('.')[0].strip()
conc = lambda x : x[:2].strip()
```

```python
def data_1 (df, col) :
    con_1 = lambda x : x + "-"+ col +"(단위 천원)"
    ab = []
    if (col=="1인당 지역내총생산") :
        for i in range(2020-2003) :
            ab.append(str(i + 2003))
    elif (col=="1인당 지역총소득") :
        for i in range(2020-2003) :
            ab.append(str(i + 2003) + ".1")
    elif (col=="1인당 개인소득") :
        for i in range(2020-2003) :
            ab.append(str(i + 2003) + ".2")
    elif (col=="1인당 민간소비") :
        for i in range(2020-2003) :
            ab.append(str(i + 2003) + ".3")
    df.rename(columns={"2019 p)":"2019","2019 p).1":"2019.1","2019 p).2":"2019.2","2019 p).3":"2019.3"}, 
             inplace=True)
    df = df[ab]
    df.columns = df.columns.map(con)
    df = df.transpose().drop(columns="시도별")
    df.rename(columns={"충청북도":"충북", "충청남도":"충남", "전라북도":"전북", 
                      "전라남도":"전남", "경상북도":"경북","경상남도":"경남"}, inplace=True)
    df = df.astype("int")
    df.columns = df.columns.map(conc).map(con_1)
    df["수도권-"+ col +"(단위 천원)"] = (df["인천-"+ col +"(단위 천원)"] + df["서울-"+ col +"(단위 천원)"] 
                      + df["경기-"+ col +"(단위 천원)"])/3
    return df
```

```python
data1 = data_1(data1,"1인당 개인소득")
data1
```

```python
def data_2 (df) :
    con_2 = lambda x : x + "-주택가격지수(2017.11월 기준)"
    df.drop(columns="주택유형별(1)", inplace=True)
    df = df.transpose()
    df.index = df.index.map(con)
    df.columns = df.columns.map(con_2)
    df["수도권-주택가격지수(2017.11월 기준)"] = (df["인천-주택가격지수(2017.11월 기준)"] 
                                     + df["서울-주택가격지수(2017.11월 기준)"] 
                                     + df["경기-주택가격지수(2017.11월 기준)"])/3
    df = df.astype("float")
    df = df.groupby(df.index).mean()
    return df
```

```python
data2 = data_2(data2)
```

```python
def data_3 (df) :
    con_3 = lambda x : x + "-인구(단위 명)"
    df = df.drop(index=0).transpose()
    df.columns = df.iloc[0]
    df.drop(index="행정구역(시군구)별", inplace=True)
    df.rename(columns={"충청북도":"충북", "충청남도":"충남", "전라북도":"전북", 
                      "전라남도":"전남", "경상북도":"경북","경상남도":"경남"}, inplace=True)
    df = df.astype("float")
    df.columns = df.columns.map(conc).map(con_3)
    df["수도권-인구(단위 명)"] = (df["인천-인구(단위 명)"] 
                                    + df["서울-인구(단위 명)"] 
                                    + df["경기-인구(단위 명)"])/3
    return df
```

```python
data3 = data_3(data3)
```

```python
data = pd.concat([data1, data2, data3], axis=1)
```

```python
data = data.round(decimals=2)
data
```

# 인구와 주택가격지수

```python
location = ['강원', '경기', '경북', '경남', '광주', '대구', '대전', '부산', '서울',
            '세종', '울산', '인천', '전남', '전북', '제주','충남', '충북']
def data_corr (df) :
    loc_value=[]
    for i in location :
        dff = df.filter(regex=i, axis=1)
        loc_value.append((dff.corr().loc[i+"-인구(단위 명)"][1]*100).round(2))
    return loc_value
```

```python
corr_value = data_corr(data)
corr_value
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import geopandas as gpd
from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False
f_path = "C:/Windows/Fonts/Hancom Gothic Regular.ttf"
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)

map_k = gpd.read_file("../../Documents/Data/map/TL_SCCO_CTPRVN.shp", encoding = 'euc-kr')
map_k
```

```python
map_k["지수&인구_상관관계"] = corr_value
map_k
```

```python
# 지도 위에 표시하기
def geowrite (a) :
    f = a.plot(cmap='Blues',column="지수&인구_상관관계", edgecolor="w", figsize=(15,15),vmin=-100,vmax=100, legend=True)
    f.set_title("\n주택가격지수 & 인구 상관관계도 구역도\n[2017.11.01 = 100]기준")
    f.set_axis_off()
    plt.show()
```
