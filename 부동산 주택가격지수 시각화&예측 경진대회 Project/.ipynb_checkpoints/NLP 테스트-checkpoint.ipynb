```python
list1 = ko.nouns(text)
type(list1)
```

```python
def getNoun(list) :
    cnt_dic = {}
    for x in list :
        if cnt_dic.get(x) :
            cnt_dic[x] += 1
        else :
            cnt_dic[x] = 1
    return cnt_dic
```

```python
getNoun(list1)
```
