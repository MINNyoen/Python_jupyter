# 주택매매가격지수

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
MR1 = pd.read_csv("../../Documents/Data/Project/주택매매가격지수(KB).csv", encoding="utf-8", engine="python", index_col=0)
MR2 = pd.read_csv("../../Documents/Data/Project/주택전세가격지수(KB).csv", encoding="utf-8", engine="python", index_col=0)
```

```python
con = lambda x : x.split('-')[0].strip()
def slic (a, b) :
    a.rename(columns={'총지수[2019.01=100]': b + '-총지수', '단독주택[2019.01=100]' : b + '-단독주택',
                   '연립주택[2019.01=100]' : b + '-연립주택','아파트[2019.01=100]':b + '-아파트', 
                   '아파트(서울)[2019.01=100]': b + '-아파트(서울)', '총지수(서울)[2019.01=100]' : b + '-총지수(서울)'},
                     inplace=True)
    a.index = a.index.map(con)
    a = a.groupby(a.index).mean()
    return a
```

```python
MR1 = slic(MR1, "매매")
MR2 = slic(MR2, "전세")
result = pd.concat([MR1,MR2], axis=1)
result
```

```python
result.loc[["1986"]]
```

```python
result.corr()
```

```python
%matplotlib inline

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

f_path = "C:/Windows/Fonts/Hancom Gothic Regular.ttf"
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)
```

```python
def plotgraph (name ,a, b) :
    if (name != None) :
        result1= result.filter(regex=name, axis=1)
    plt.figure(figsize=(20,10))
    plt.plot(result1.loc[a:b].index, result1.loc[a:b], label = result1.columns)
    plt.xlabel("연도", fontsize=20)
    plt.ylabel("주택매매-전세가격지수\n(2019.01기준 = 100)", fontsize=15)
    plt.legend(loc = "upper left")
    plt.show
```

```python
#1986년 부터 2021까지
plotgraph("전세","1986", "2021")
```

# 인사이트 정리
## 1. 1986년부터 2021년까지 지속적으로 증가하는 모습을 보인다.
## 2. 1997-1998년도, 2001-2005년까지 모든 가격지수가 낮아지는 모습을 보인다.(이유)
## 3. 1986년부터 2021년까지 단독주택(전세)은 가장 낮은 성장률을, 아파트(전세)는 가장 높은 성장률을 보인다.

```python
np.arange(0,101,10)
```

```python
def barchart (year,name = "전체") :
    result1 = result
    if (name == "전체") :  
        result1 = result.loc[[year]]
    else :
        result1 = result.filter(regex=name, axis=1).loc[[year]]
    labels = result1.columns
    x = np.arange(len(labels))
    y = np.arange(0,131,10)
    plt.figure(figsize=(20,10))
    plt.bar(labels, result1.loc[year].values, width=0.3, label = year + "년도"+ name +"관련"+ "주택가격 지수")
    plt.xlabel("검색한 데이터", fontsize=20)
    plt.ylabel("주택매매-전세가격지수\n(2019.01기준 = 100)", fontsize=15)
    plt.xticks(labels)
    plt.yticks(y)
    plt.legend(loc = "upper left", fontsize=20)
    plt.show
```

```python
barchart("2021")
```

```python
def barchart2 (year1,year2,name = "전체") :
    result1 = result
    if (name == "전체") :  
        result1 = result.loc[[year1,year2]]
    else :
        result1 = result.filter(regex=name, axis=1).loc[[year1,year2]]
    labels = result1.columns
    x = np.arange(len(labels))
    y = np.arange(0,131,10)
    width=0.3
    plt.figure(figsize=(20,10))
    plt.bar(x-width/2, result1.loc[year1].values, width=width, label = year1 + "년도"+ name +"관련"+ "주택가격 지수")
    plt.bar(x+width/2, result1.loc[year2].values, width=width, label = year2 + "년도"+ name +"관련"+ "주택가격 지수")
    plt.xlabel("검색한 데이터", fontsize=20)
    plt.ylabel("주택매매-전세가격지수\n(2019.01기준 = 100)", fontsize=15)
    plt.xticks(ticks=x, labels=labels)
    plt.yticks(y)
    plt.legend(loc = "upper left", fontsize=20)
    plt.show
```

```python
barchart2("1989","2021")
```
