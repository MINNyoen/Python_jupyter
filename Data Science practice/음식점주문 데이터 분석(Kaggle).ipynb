## 주문당 평균 계산금액 출력하기

```python
food1 =  pd.read_csv('../../Documents/Data/chipotle.tsv', sep='\t')
food1
```

```python
#평균 계산금액이기 때문에 float으로 변경해야함
result = []
for i in range(len(food1.index)):
    result.append(float(food['item_price'][i][1:]))
print(result[1])
```

```python
food1.insert(column="금액($)", loc=4, value=result)
food1['금액($)']
```

```python
food.groupby("order_id").mean()
```

```python
foodorder = food.groupby("order_id").mean().reset_index().sort_values(by=['금액($)'], ascending = False)
foodorder
```

## 한 주문당 40달러 이상의 주문의 id 출력하기
