# Data Practice

```python
dp = pd.read_csv("../../Documents/Data/world-happiness-report-2021.csv", encoding='utf-8')
dp
```

```python
dp_east = dp[dp["Regional indicator"]=="East Asia"].copy()
dp_east
```

```python
dp_east_ladder= dp_east.sort_values(by='Ladder score', ascending=False).iloc[0:5]
dp_east_ladder
```

```python
dp_east_life= dp_east.sort_values(by='Healthy life expectancy', ascending=False).iloc[0:5]
dp_east_life
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

f_path = "C:/Windows/Fonts/Hancom Gothic Regular.ttf"
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)
```

```python
labels = dp_east_life['Country name']
x = np.arange(len(labels))  
width = 0.35  

fig, ax = plt.subplots(figsize=(16,8))
rects1 = ax.bar(x - width/2, dp_east_life['Ladder score'], width, label='행복지수', color='red')
rects2 = ax.bar(x + width/2, dp_east_life['Healthy life expectancy'], width, label='기대수명')

ax.set_ylabel('기대수명 : age, 행복지수 : %')
ax.set_title('동아시아국가의 기대수명에대한 행복지수 그래프')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
