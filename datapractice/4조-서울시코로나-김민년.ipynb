```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vc = pd.read_csv('../../Documents/Data/서울특별시 코로나19 백신 예방접종 현황.csv', engine="python")
vc
```

```python
vc.columns
```

```python
vcf = vc[['기준일', '1분기 1차접종대상자(서울시)','1분기 1차실접종자(서울시)', '1분기 1차접종률(서울시)','1분기 2차접종대상자(서울시)',
          '1분기 2차실접종자(서울시)', '1분기 2차접종률(서울시)', '1분기 1차 추가접종자(서울시)','1분기 2차 추가접종자(서울시)']]
vcf
```

```python
cr = lambda x : x.split('.')[1].strip()
vcf['기준일'] = vcf['기준일'].map(cr)
vcf
```

```python
vcf = vcf.drop(['기준월'], axis=1)
vcf
```

```python
vcf.rename(columns= {'기준일':'월'}, inplace=True)
vcf
```

```python
vcfa = vcf.groupby('월').mean()
```

```python
vcfm = vcf.groupby('월').sum()
vcfm =vcfm.reset_index()
vcfm
```

```python
%matplotlib inline

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

f_path = "C:/Windows/Fonts/Hancom Gothic Regular.ttf"
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)

labels = vcfm['월']
x = np.arange(len(labels))  
width = 0.3

fig, ax = plt.subplots(1,2, figsize=(14,8))
ax[0].bar(x - width/2, vcfm['1분기 1차접종대상자(서울시)'], width, label='접종대상자', color='red')
ax[0].bar(x + width/2, vcfm['1분기 1차실접종자(서울시)'], width, label='실접종자')

ax[1].bar(x - width/2, vcfm['1분기 2차접종대상자(서울시)'], width, label='접종대상자', color='orange')
ax[1].bar(x + width/2, vcfm['1분기 2차실접종자(서울시)'], width, label='실접종자', color='green')

ax[0].set_ylabel('빨강 : 백만명 기준, 파랑 : 백만명 기준')
ax[0].set_xlabel('월')
ax[0].set_title('서울시 코로나백신 예방접종 현황 (2,3,4월)')
ax[0].set_xticks(x)
ax[0].set_xticklabels(labels)
ax[0].legend()

ax[1].set_ylabel('빨강 : 백만명 기준, 파랑 : 백만명 기준')
ax[1].set_xlabel('월')
ax[1].set_title('서울시 코로나백신 예방접종 현황 (2,3,4월)')
ax[1].set_xticks(x)
ax[1].set_xticklabels(labels)
ax[1].legend()



plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(14,8))

ax[0].bar(x, vcfa['1분기 1차접종률(서울시)'], width, label='1차접종률', color='red')
ax[1].bar(x, vcfa['1분기 2차접종률(서울시)'], width, label='2차접종률', color='blue')


ax[0].set_ylabel('접종률 : %')
ax[0].set_xlabel('월')
ax[0].set_title('서울시 코로나백신 예방접종률 현황 (2,3,4월)')
ax[0].set_xticks(x)
ax[0].set_xticklabels(labels)
ax[0].legend()

ax[1].set_ylabel('접종률 : %')
ax[1].set_xlabel('월')
ax[1].set_title('서울시 코로나백신 예방접종률 현황 (2,3,4월)')
ax[1].set_xticks(x)
ax[1].set_xticklabels(labels)
ax[1].legend()



plt.show()
```
