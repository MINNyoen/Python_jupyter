```python
Rare = [ 'Col', 'Mme', 'Dr', 'Major', 'Capt',
       'Lady', 'Sir', 'Mlle', 'Dona', 'Jonkheer', 'the Countess', 'Don',
       'Rev', 'Ms']
for x in Rare:
    td['title'] = td['title'].replace(x, 'Rare')
    
td['title'].unique()
#Rare에 있는 리스트들을 다 Rare로 변환하기
```

```python
tmp2 = []
for x in td['sex']:
    if x == 'female':
        tmp2.append(0)
    elif x == 'male':
        tmp2.append(1)
    else:
        tmp2.append(np.nan)
len(tmp2)
td['sex'] = tmp2
```

```python
td['parch'] = td['parch'].astype('float')
td['survived'] = td['survived'].astype('float')
td['pclass'] = td['pclass'].astype('float')
td['sex'] = td['sex'].astype('float')
td['sibsp'] = td['sibsp'].astype('float')
td['parch'] = td['parch'].astype('float')
td['fare'] = td['fare'].astype('float')
td.head()
```

```python
td = td[td['age'].notnull()]
td = td[td['sibsp'].notnull()]
td = td[td['parch'].notnull()]
td = td[td['fare'].notnull()]
td.info()
```

```python
train_pre = td[['pclass','sex','age','sibsp','parch','fare']]
train_pre
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_pre, #데이터 프레임
                 td[['survived']], #맞출 정답
                 test_size = 0.1, #10퍼 테스트 데이터 10%, 훈련데이터 90% 할당
                 random_state = 13 #Seed값
                ) 
```

```python
X_test = X_test.reset_index()
y_test = y_test.reset_index()
y_train = y_train.reset_index()
```

```python

X_test = X_test.drop(['level_0'], axis=1)

y_train = y_train.drop(['level_0'], axis=1)

y_test = y_test.drop(['level_0'], axis=1)
```

```python
X_train = X_train.drop(['level_0'], axis=1)
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.multioutput import MultiOutputClassifier
tree_clf = DecisionTreeClassifier(max_depth = 3, random_state = 13)
tree_clf.fit(X_train, y_train)
```

```python
tree_clf.score(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = tree_clf.predict(X_test)
accuracy_score(y_test, y_pred)
```
