# 1. 종합 - 지역별(전국-수도권-지방)

```python
MR
con = lambda x : x.split('-')[0].strip()
MR.index = MR.index.map(con)
```

```python
MR_m = MR.groupby(MR.index).mean().filter(regex='^종합', axis=1)
MR_m
```

```python
MR_m.columns[0:3][0:3]
```

```python
%matplotlib inline

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

f_path = "C:/Windows/Fonts/Hancom Gothic Regular.ttf"
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)
```

```python
plt.plot(figsize=(20,8))
plt.plot(MR_m.index, MR_m.iloc[0:,0:1].values, label = "전국")
plt.plot(MR_m.index, MR_m.iloc[0:,1:2].values, label = "수도권")
plt.plot(MR_m.index, MR_m.iloc[0:,2:3].values, label = "지방")

plt.xlabel("연도")
plt.ylabel("주택가격지수\n(2017.11기준 = 100)")
plt.legend()
plt.show
```
